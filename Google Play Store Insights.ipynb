**Overview**

*What is EDA?*

* EDA (Exploratory Data Analysis) is an approach of analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visualization methods.

*Why we need EDA?*

* EDA help us look at data before making any assumptions. It can help identify obvious errors, as well as better understand patterns within the data, detect outliers or anomalous events, find interesting relations among the variables.
* Data scientists can use exploratory analysis to ensure the results they produce are valid and applicable to any desired business outcomes and goals.

*How to do EDA?*

* In EDA we will follow below steps

  i) Knowing our data -> Understanding the dataset size and shape

  ii)  Understanding our variables -> Exploring each variable to know the distribution, how it categorized and so on

  iii) Data Wrangling -> This is the process of removing errors and combining complex data sets to make them more accessible and easier to analyze.

  iv) Data Vizualization -> This is the representation of data through use of common graphics, such as charts, plots. These visual displays of information communicate complex data relationships and data-driven insights in a way that is easy to understand.

  v) Solution - Providing solutions by investigating the data relationships and data-driven insights

## 2. **Data Wrangling**

### Data Wrangling Code

```python
# Data Set information before wrangling
df.info()
```

```python
#Dropping Unecessary Columns

df.drop(['Size','Genres','Android Ver','Current Ver'],axis=1,inplace=True)
```

```python
# Dropping Missing Values
df.dropna(subset=['Content Rating','Type'],inplace=True)
df.reset_index(drop=True,inplace= True)
```

```python
# Interpolation on Ratings Columns

# Grouping the data according to the 'Installs' category by applying median function to the 'Rating' category
# After grouping, stored indexes of the Pandas series in a list (index_list)
# After grouping, stored values of the Pandas series in a list (values_list)

index_list = list(df.groupby(['Installs'])['Rating'].median().index)
values_list = list(df.groupby(['Installs'])['Rating'].median().values)

# Using comprehension, created a dictionary of index_list & values_list

rating_dictionary = {index_list[i]:values_list[i] for i in range(len(index_list))}
```

```python
# Defining a function to return the respective value of the dictionary 'rating_dictionary' for the given key

def matching_function(y):
  return rating_dictionary[y]
```

```python
# Lambda function will check each row if the 'Rating' value is NaN.
# If its true then it will apply the 'matching_function' function else it will return the 'Rating' Value as the same

df['Rating'] = df.apply(lambda x:matching_function(x['Installs']) if pd.isnull(x['Rating']) == True else x['Rating'],axis=1)
```

```python
# Droppig the rows where 'Rating' & 'Installs' values are zero

df.dropna(subset = ['Rating'],inplace=True)
```

```python
# Changing data types
df['Reviews'] = df['Reviews'].astype(int)
df['Price'] = df['Price'].str.replace(r'$', '', regex=True).astype(float)
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
```

```python
# Extracting Day,Date & Year

df['Last Updated Years'] = df['Last Updated'].dt.year
df['Last Updated Months'] = df['Last Updated'].dt.strftime('%B')
df['Last Updated Days'] = df['Last Updated'].dt.day
```

```python
# Adding a column 'Installs (Int)'
df['Installs (Int)']=df['Installs'].str.replace(r'\D', '', regex=True).astype(int)
```

```python
df = df.sort_values(by='Last Updated',ascending = False).drop_duplicates(subset='App',ignore_index=True)
```
